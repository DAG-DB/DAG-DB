```python
# Draw DAGs (and a not-DAG) for thesis Ch 1, and for app A.2 Reducing variance
# patterns in linear additive noise models


import matplotlib.pyplot as plt
import networkx as nx
```

```python
def get_digraph(edges, nodes=None):
    g = nx.DiGraph()
    if nodes is not None:
        g.add_nodes_from(nodes)
    g.add_edges_from(edges)
    return g


def draw_g(g, edges, file_title=None, figsize=(5, 5), edge_color='k'):
    fig = plt.figure(figsize=figsize)
    nx.draw_networkx(
        g, node_size=1200, font_size=20, font_color='k', arrowsize=40,
        pos = {0: (0, 0), 1: (1, 2), 2: (2, 0), 3: (3, 2)},
        edgelist=edges, edge_color=edge_color, linewidths=2
    )
    if file_title is not None:
        fig.savefig(f'{file_title}_drawing.pdf')


def draw(edges, nodes=None, file_title=None, figsize=(5, 5), edge_color='k'):
    g = get_digraph(edges, nodes=nodes)
    draw_g(
        g, edges, file_title=file_title, figsize=figsize,
        edge_color=edge_color)
```

```python
draw([(0, 1), (0, 2), (1, 3)], file_title='dag_1')
```

```python
draw([(0, 1), (0, 2), (0, 3), (1, 3)], file_title='dag_2')
```

```python
draw(
    [(0, 1), (0, 2), (3, 0), (1, 3)], file_title='not_dag',
    edge_color=['b', 'k', 'b', 'b'])
```

```python
draw([(0, 1), (0, 2), (1, 3), (2, 3)], file_title='dag_cov')
```
