# Example of how to run DAB-DB

This simply runs `main_dag.py` from the linux command line, with a
configuration file passed in as an argument

```python
!python3 main_dag.py -c 'imle_logistic_none'
```

This runs 24 DAGs, each for 1000 epochs, so will take a considerable time to
complete.  The output will be logged in `../logs1 under the run name indicated
at the start of the above output

Could instead pass to the `-c` argument, for example, any of
`ste_logistic_none`, `ste_logistic_84`, `imle_logistic_66`,
`sfe_logistic_66`.
