```python
import numpy as np
import os
from data_management.synthetic_dataset import SyntheticDataset


def save_dag(X, B, n, i):
    folder = os.path.join('..','..','..','data', 'big', str(n).zfill(3))
    # print(os.path.abspath(folder))
    # return
    if not os.path.exists(folder):
        os.mkdir(folder)
    file_stem = os.path.join(folder, f'{str(i).zfill(2)}_')
    np.save(file_stem + 'data.npy', X)
    np.save(file_stem + 'adj.npy', B)


def generate_dag_data(n, i, seed):
	data = SyntheticDataset(
		1200, 30, 'ER', 2 * 2, 'gaussian_ev', 1., seed=seed)
	save_dag(data.X, data.B, n, i)


def create_save(n, seed_list):
	for i, seed in enumerate(seed_list):
		assert isinstance(seed, int)
		generate_dag_data(n, i, seed)
```

```python
for n, seed_list in enumerate(seed_bank):
    create_save(n, seed_list)
```
