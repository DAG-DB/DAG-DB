#### Calculate and show relative sizes


```python
size_trues = dict()
for dag_folder in os.listdir(DATA_FOLDER):
    dag_filepath = os.path.join(DATA_FOLDER, dag_folder)
    if not os.path.isdir(dag_filepath):
        continue
    graph_set = dag_folder
    size_trues_set = - np.ones(N_DAGS)
    for adj_true_file in os.listdir(dag_filepath):
        graph_type = dag_folder[3:6]
        d = int(dag_folder[:3])
        k = int(dag_folder[6])
        if ('data' in adj_true_file) or (adj_true_file[-4:] == '.log'):
            continue
        graph_no = int(adj_true_file[: 2])
        size_trues_set[graph_no] = np.load(
            os.path.join(dag_filepath, adj_true_file)).astype(
            bool).astype(int).sum()
    assert np.all(size_trues_set != -1)
    size_trues[graph_set] = size_trues_set
```

```python
def get_rel_size(df_row):
    size_trues_set = size_trues[df_row['graph_set']]
    size_pred = df_row['size']
    rel_size = size_pred / size_trues_set[int(df_row['graph_no'])]
    return rel_size


results['rel_size'] = results.apply(get_rel_size, axis=1)
```

```python
g = sns.catplot(
    data=results, x='algo', y='rel_size',
    row='graph_type', col='d', kind='box',
    showmeans=True, meanprops=meanprops,
    sharey='row'
)
for ax in g.axes.flat:
    labels = ax.get_xticklabels()
    ax.set_xticklabels(labels, rotation = 75, ha='right')
g.set(xlabel=None, ylabel=None)
# plt.ylim(0, 1)
plt.tight_layout()  # Avoid labels cut off in save
plt.savefig('vmt_rel_size.pdf');
```
